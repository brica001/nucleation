```python
from scipy.integrate import odeint

q=72 #number of surface proteins
Ccmc=0.2 #critical concentration, micrometer
alpha=2 #line energy kT 
beta=70 #rate of attachment, Hz/micrometer
c1t0=10 #initial concentration, micrometers
Rh=0.619 #hydrodynamic radius, micrometer, estimated for 300 amino acid protein
R=np.sqrt(q)*Rh/2 #estimate for spherical capsid radius
T=22 #room temp, celcius
k=1.38064852*10**(-23) #boltzmann constant
Z= 0.9544 #viscosity room temp water, millipascals
D1=k*T/(6*np.pi)*Rh*Z
j=D1*2*R/q
def zeros(p): #makes list of p zeros
    return [0] * p
init= zeros(q)
init[1]=c1t0
def Con(cn,t):
    rates= zeros(q)
    rates[1]=-(cn[1]**2)*j*np.sqrt(q-1)+cn[2]*cn[1]*j*np.sqrt(q-1)*np.exp(alpha*(np.sqrt(2*q-4)+np.sqrt(q-1)))/(cn[1]*2/Ccmc)
    for n in (2,q-1):
        bn=cn[1]*j*sqrt(n*(q-n))#unbinding energy for capsid size n
        bnm1=cn[1]*j*sqrt((n-1)*(q-n+1))#unbinding energy for capsid size n+1
        Gn=n*np.log(cn[1]*2/Ccmc)+alpha*np.sqrt(n*(q-n))#free energy for capsid size n
        Gnp1=(n+1)*np.log(cn[1]*2/Ccmc)+alpha*np.sqrt((n+1)*(q-n-1))#free energy for capsid size n+1
        Gnm1=(n-1)*np.log(cn[1]*2/Ccmc)+alpha*np.sqrt((n-1)*(q-n+1))
        alpn=bnm1*np.exp(Gn-Gnm1)
        alpnp1=bnm1*np.exp(Gnp1-Gn)
        rates[n]=bnm1*cn[n-1]-alpn*cn[n]-bn*cn[n]+alpnp1*cn[n+1]
        return(rates)
t=np.linspace(0,120,120)       
x=odeint(Con,init,t)    
   



plt.plot(t,x[q])
```
